```python
import pandas as pd
import googletrans 
from googletrans import Translator
```

```python
import pandas as pd
```

```python
# Load dataset
def load_data():
    data = pd.read_csv("FIX2023-SK.csv")#ubah nama file sesai dengan nama file kalian
    return data
```

```python
df = load_data()
df.head()
```

```python
translator = Translator()
translations = {}
for column in df.columns:
    #unique elements of the column
    unique_elements = df[column].unique()
    for element in unique_elements:
        #adding all the translations to a dictionary
        translations[element] = translator.translate(element).text
translations
```

```python
df.replace(translations, inplace = True)
df.head(10)
```

```python
from textblob import TextBlob #NLP bahasa inggris
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
%matplotlib inline
```

```python
def getSubjectivity(text): 
    return TextBlob(str(text)).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(str(text)).sentiment.polarity

df['Subjectivity'] = df['tweet'].apply(getSubjectivity)
df['Polarity'] = df['tweet'].apply(getPolarity)

df
```

```python
def getSentiment(score):
    if score < 0 :
        return 'negatif'
    elif score == 0 :
        return 'netral'
    else :
        return 'positif'
# Apply the function
df['Sentiment'] = df['Polarity'].apply(getSentiment)
#lets take a look
df
```

```python
def getSentiments(kata):
    if kata == 'negatif' :
        return '-1'
    elif kata == 'netral' :
        return '0'
    else :
        return '1'
# Apply the function
df['Sentiments'] = df['Sentiment'].apply(getSentiments)
#lets take a look
df
```
